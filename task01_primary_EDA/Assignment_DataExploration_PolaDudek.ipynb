## Exercise 2
Using scatter plot:  
a) Show the relation of square footage and price.  
b) Check how price is influenced by apartment's grade. Find other variables correlated with price.  
c) Check how good apartments are distributed over the city. First, plot zipcode versus price. Then, use apartments coordinates (longitude and lattitude) in order to show where apartments of good grade are located.

```python
#a)
#Your code:
x = kc_house_data['sqft_living']
y = kc_house_data['price']
plt.scatter(x, y)
plt.xlabel('Living area [sqft]')
plt.ylabel('Price')
plt.show()

#Hint plt.scatter
```

**Note:** Scatterplot allows us to visually represent the relationship between house price and living area in square feet. As expected, the price generally increases with the living area. However it is not a "thin line" type of correlation - there is a large dispersion of prices for each value of *sqft_living*.

```python
#b)
#Your code:
x = kc_house_data['grade']
y = kc_house_data['price']
plt.scatter(x, y)
plt.xlabel('Grade')
plt.ylabel('Price')
plt.show()
```

**Note:** As house grade increases, the maximum price increases too. However, the lowest prices change much more slowly, which means that is is possible to find "bargain houses" even in the higher grades.

```python
x = kc_house_data["condition"]
y = kc_house_data["price"]
plt.scatter(x, y)
plt.xlabel("Condition")
plt.ylabel("Price")
plt.show()
```

**Note:** The highest prices are found in condition "3" rather than in the higher condition "5" - it is difficult to establish a reasonable relationship between these two variables.

```python
x = kc_house_data["yr_built"]
y = kc_house_data["price"]
plt.scatter(x, y)
plt.xlabel("Year built")
plt.ylabel("Price")
plt.show()
```

**Note:** For the year built and price, it is impossible to estabish a clear correlation - for each year, the range of prices is large and relatively similar.

```python
x = kc_house_data["bathrooms"]
y = kc_house_data["price"]
plt.scatter(x, y)
plt.xlabel("Number of bathrooms")
plt.ylabel("Price")
plt.show()
```

**Note:** Correlation between the number of bathrooms and price is quite visible - the higher the number of bathrooms, the higher the maximum found price. However, similar to the *grade* and *price* plot, the lowest prices remain fairly constant.

```python
#c)
#Your code:
x = kc_house_data['zipcode']
y = kc_house_data['price']
plt.scatter(x, y)
plt.xlabel('Zipcode')
plt.ylabel('Price')
plt.show()
```

```python
#Your code:
x = kc_house_data['long']
y = kc_house_data['lat']
scatter_plt = plt.scatter(x, y, c=kc_house_data['grade'])
cbar = plt.colorbar(scatter_plt)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

**Note:** By basing the color of each point on its house grade, we are able to determine that the "best" houses are located
between 47.5 and 47.7 latitude and -122.3 and -121.9 longitude.

## Exercise 3
a) Create boxplot of apartment's grade versus it's price. Compare with plot 2b).  
b) Create mosaic plot of any two categorical variables.

```python
#a)
#Your code:
grades_with_prices = kc_house_data.filter(items=["grade", "price"])
grades_with_prices.boxplot(column='price', by='grade')
plt.xlabel('Grade')
plt.ylabel('Price')
plt.show()
#Hint: plt.boxplot
```

**Note:** Boxplot allows us to see the middle 50% of house price range for each grade *(the box)*, alongside with the rest of the distribution *(dots)*. Compared to *2b)* plot it provides a clearer view of how the density of prices is distributed.

```python
#b)
from statsmodels.graphics.mosaicplot import mosaic

#Your code:
mosaic(kc_house_data, ['view', 'grade'])
plt.xlabel('View')
plt.ylabel('Grade')
plt.show()
```

**Note:** Mosaic plot represents the number of combinations between values of 2 given categorical variables through the size of their rectangles. For example, the most common combination of grade and view values is *grade = 7* and *view = 0*.

## Exercise 4
"I am interested in the house with 3 bedrooms, 2 bathrooms and 2 floors"  
a) Compute the basic statistics of apartments fulfilling above constraints.  
b) Distinct accepted apartments on any of previously created plots.  
c) Find such apartments with best price to footage ratio.  

```python
#a)
#Your code:
found_houses = kc_house_data.loc[
    (kc_house_data['bedrooms'] == 3) &
    (kc_house_data['bathrooms'] == 2) &
    (kc_house_data['floors'] == 2)
].copy() # using copy() to create a new dataframe instead of a "view" inside the original

found_houses.head()
#Hint: pandas.DataFrame.loc
```

```python
#b)
#Your code:
plt.scatter(kc_house_data['grade'], kc_house_data['price'])
plt.scatter(found_houses['grade'], found_houses['price'])
plt.xlabel('Grade')
plt.ylabel('Price')
plt.show()
```

```python
#c)
#Your code:
found_houses['price_to_footage_ratio'] = found_houses['price'] / found_houses['sqft_living']
found_houses.sort_values(by='price_to_footage_ratio', ascending=True, inplace=True)
best_houses_idxs = found_houses.index[:5]
best_houses = found_houses.loc[best_houses_idxs]
best_houses


#Hint: .loc, .sort_values(), .index
```

**Note:** The best ratio is when the price is the smallest and sqfr_living is the biggest - the best ratio is therefore the smallest one. Using `index()` function I selected 5 houses with the smallest ratio - the *"best"* ones fitting given criteria.
